```python
learn=0

while learn < 40:
  index = 0
  learn +=1
  for Input in inputs:
    
    #Neurona_1
    Sum = Input[0] * weight1[0] \
        + Input[1] * weight1[1] \
        + Input[2] * weight1[2] \
        + Input[3] * weight1[3] \
        + Input[4] * weight1[4] \
        + Input[5] * weight1[5] \
        + Input[6] * weight1[6] + weight1[7]
        
    E1=salida[index]-Sum

    weight1[0]+=LRate*E1*Input[0]
    weight1[1]+=LRate*E1*Input[1]
    weight1[2]+=LRate*E1*Input[2]
    weight1[3]+=LRate*E1*Input[3]
    weight1[4]+=LRate*E1*Input[4]
    weight1[5]+=LRate*E1*Input[5]
    weight1[6]+=LRate*E1*Input[6]
    weight1[7]+=LRate*E1
    
    index += 1

```
