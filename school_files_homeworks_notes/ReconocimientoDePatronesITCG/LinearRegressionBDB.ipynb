```python
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

#Simulación de los datos
def funcionLineal(x):
  np.random.seed(0)
  ruido = 0.2*np.random.randn(x.shape[0])
  b= 3.25 + ruido
  m = 0.1
  y = m*x + b
  return y

x = np.arange(0, 30, 0.5) 
y = funcionLineal(x) 

plt.scatter(x,y,label='data', color='blue')
plt.title('Datos');
```

```python
# Importamos la clase de Regresión Lineal de scikit-learn
from sklearn.linear_model import LinearRegression 
regresion_lineal = LinearRegression()

#Entrenamiento
regresion_lineal.fit(x.reshape(-1,1), y) 
#Coeficientes
w=regresion_lineal.coef_
b=regresion_lineal.intercept_
print('w = ',w[0], ' b = ',b)
```

```python
#Función lineal con los valores del modelo
def funcionLineal(x,w,b):
    y = w[0]*x + b
    return y

yy = funcionLineal(x,w,b)

plt.plot(x,yy,label='data', color='red')
plt.scatter(x,y,label='data', color='blue')

plt.title('Datos');
plt.show()
```

```python
#Error cuadrático medio (MSE)
from sklearn.metrics import mean_squared_error
# Predecimos los valores y para los datos usados en el entrenamiento
y_Pred = regresion_lineal.predict(x.reshape(-1,1))

mse = mean_squared_error(y_true = y, y_pred = y_Pred)
# La raíz cuadrada del MSE es el RMSE
rmse = np.sqrt(mse)
print('Error Cuadrático Medio (MSE) = ' + str(mse))
print('Raíz del Error Cuadrático Medio (RMSE) = ' + str(rmse))
```

```python
#Coeficiente de determinación R2
r2 = regresion_lineal.score(x.reshape(-1,1), y)
print('Coeficiente de Determinación R2 = ' + str(r2))
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score


a = datasets.load_diabetes
print(a)
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)
diabetes_X = diabetes_X[:, np.newaxis, 2]


diabetes_X_train = diabetes_X[:-20]
diabetes_X_test = diabetes_X[-20:]

diabetes_y_train = diabetes_y[:-20]
diabetes_y_test = diabetes_y[-20:]

regr = linear_model.LinearRegression()

regr.fit(diabetes_X_train, diabetes_y_train)

diabetes_y_pred = regr.predict(diabetes_X_test)

# The coefficients
print("Coefficients: \n", regr.coef_)
# The mean squared error
print("Mean squared error: %.2f" % mean_squared_error(diabetes_y_test, diabetes_y_pred))
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(diabetes_y_test, diabetes_y_pred))

plt.scatter(diabetes_X_test, diabetes_y_test, color="black")
plt.plot(diabetes_X_test, diabetes_y_pred, color="blue", linewidth=3)


plt.xticks(())
plt.yticks(())

plt.show()
```
