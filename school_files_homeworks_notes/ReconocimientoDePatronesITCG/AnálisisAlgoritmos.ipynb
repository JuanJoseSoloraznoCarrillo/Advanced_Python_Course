```
#EJERCICIO 3.1-2 DOMOSTRAR C1*n <= (n-a)

for n in range(1,41):
  for a in range (-40,41):
    C1 = -abs(a)
    if (C1*n <= (n + a)):
      print("cumple con n = ",n,"y a = ",a)
      #print("C1 = ",C1)
      #print((C1*n)," = ",(n+a))
    else:
      print("posible no cumpla por: ",C1*n,"!=",(n+a))
      CC1 = round(C1*n)
      if (CC1 <= (n + a)):
        print("cumple con n = ",n)
      else:
        print("\n¡NO CUMPLE CON n = ",n,"y C1 = ",C1,"a = ",a,"\n\n\n")
   

```

```
#EJERCICIO 3.1-2 DOMOSTRAR C1*n >= (n+a)

for n in range(1,41):
  for a in range (-40,41):
    C2 = abs(a)+1
    if (C2*n >= (n + a)):
      print("cumple con n = ",n,"y a = ",a)
      #print("C2 = ",C2)
      #print((C2*n)," = ",(n+a))
    else:
      print("posible no cumpla por: ",C2*n,"!=",(n+a))
      CC2 = round(C2*n)
      if (CC2 >= (n + a)):
        print("cumple con n = ",n)
      else:
        print("\n¡NO CUMPLE CON n = ",n,"y C1 = ",C2,"a = ",a,"\n\n\n")
```

```
#INSERTION SORT 
A=[31,41,59,26,41,58]

for j in range(1,len(A)):
 
  key = A[j]
  i = j - 1

  while (i>=0) and (A[i]>key):
    A[i+1] = A[i]
    i = i - 1
  A[i+1] = key

print(A)
```
