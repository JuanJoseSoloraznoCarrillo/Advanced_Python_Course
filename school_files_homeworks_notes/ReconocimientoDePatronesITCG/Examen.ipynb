```python
#Emanen
 
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.formula.api as sn
from sklearn.model_selection import train_test_split
 
#Cargar doc
df= pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Carseats.csv') # Carga el archivo .csv
df = df.loc[:,['Sales','Price','Urban','US']] # dataset de las variables que se utilizarán
Urban = df['Urban'].factorize()[0]
US = df['US'].factorize()[0]
df['US'] = US
df['Urban'] = Urban
 
model = sn.ols(formula = 'Sales ~ Price + Urban + US',data = df).fit()
print(model.summary())
df
```
