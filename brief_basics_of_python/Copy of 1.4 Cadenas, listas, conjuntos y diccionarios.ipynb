**Tipos específicos**

Puede usarse {posición:conversión}, donde posición es el índice del argumento, y conversión se refiere al tipo de dato <br>

Símbolo  | Tipo de dato
---------|------------------
s    | Cadenas 
d    | Entero decimal
f    | Decimal

````python
grupo = "Somos {:<8} en el grupo." #Alineación izquierda
print (grupo.format (50))
grupo = "Somos {:>8} en el grupo." #Alineación derecha
print (grupo.format (49))
grupo = "Somos {:^8} en el grupo." #Alineación centrada
print (grupo.format (48))
grupo = "Somos {:=8} en el grupo." #Indica resultado positivo o negativo
print (grupo.format (47))
grupo = "Somos {:=8} en el grupo." #Indica resultado positivo o negativo
print (grupo.format (-47))
grupo = "Somos {:+8} en el grupo."
print (grupo.format (46))
grupo = "Somos {:-8} en el grupo."#Pone signo negativo
print (grupo.format (45))
grupo = "Somos {:-8} en el grupo."
print (grupo.format (-45))
grupo = "Somos {:,} en el grupo." #Separación entre miles
print (grupo.format (1234567))
grupo = "Somos {:_} en el grupo." #Separación entre miles
print (grupo.format (123456789))
````

## Modificación, inserción y borrado de elementos

En muchas ocasiones las listas necesitan insertar y borrar elementos durante la ejecución. </br>
Por ejemplo, un juego donde los jugadores disparan a los aliens (*borrar*) que caen del cielo (*insertar*) por lo que se requiere de actualizar la lista de aliens. La sintaxis para modificar es la misma que para accesar: </br>


> nombreLista[índice] = nuevoValor

Para **insertar** nuevos elementos a una lista

* Al final (añadir)
> nombreLista.append (objeto)
* En cualquier posición
> nombreLista.insert (índice, objeto)

Para **borrar** elementos de una lista
* Por posición, se debe conocer

> nombreLista.remove (objeto)

# **Organización de listas**

En muchas ocasiones no se puede predecir el orden durante la creación o mantenimiento de la lista, pero cuando se presenta o se procesa a veces es conveniente tener un orden. </br>
Algunas funciones: </br>

*   sort (), ordena de manera permanente 
*   sorted (), ordena temporalmente
*   reverse (), voltea
*   len (), número de elementos que contiene



sort () modificación permanente

````python
#Creación de una lista de cadenas
juguetes = ["muñeca", "oso", "carro", "ruleta", "rompecabezas", "casa"]
print ("Arreglo original: ", juguetes)
#ordena de forma creciente lexicográficamente, de manera permanente
juguetes.sort ()
print ("Arreglo ordenado de forma ascendente: ", juguetes)
#Orden lexicográfico decreciente
juguetes.sort (reverse=True)
print ("Arreglo ordenado en forma descendente: ", juguetes)
````

````python
#Creación de una lista de enteros
números = [3, 5, 1, 7, 8, 3, 5, 7, 2, 4];
print ("Arreglo original: ", números)
#ordena de forma no descendente 
números.sort ()
print ("Arreglo ordenado de forma no descendente: ", números)
#orden no creciente
números.sort (reverse=True)
print ("Arreglo ordenado en forma no creciente: ", números)
````

sorted () orden temporal

````python
#Para ordenar la lista de manera temporal
juguetes = ["muñeca", "oso", "carro", "ruleta", "rompecabezas", "casa"]
print ("Arreglo original: ", juguetes)
print ("Arreglo ordenado: ", sorted (juguetes))
print ("Arreglo: ", juguetes)
print ("Arreglo ordenado descendente: ", sorted (juguetes, reverse=True))
print ("Arreglo: ", juguetes)
````

reverse () para poner una lista al revés

````python
#Orden cronológico
print ("Orden cronológico: " , juguetes)
juguetes.reverse ()
print ("Después de reverse (): " , juguetes)
````

len () número de elementos de la lista

````python
#Calcula el número de elementos de una lista
print ("Lista de números: ", números)
print ("La longitud de la lista es: ", len (números))
````

# **List comprehensions**

Característica de Python. <p>

```
Lista = [expresion for item in iterable (if condición)]
```



````python
cuadrados = [n**2 for n in range (10)]
cuadrados
````

````python
#Sin list comprehensions
cuadrados = []
for n in range (10):
  cuadrados.append(n**2)
cuadrados
````

Se puede añadir una condición

````python
num = [8,1,3,2,7,0,6]
mayoresA5 = [x for x in num if x > 5]
mayoresA5
````

````python
planetas = ['Mercurio', 'Venus', 'Tierra', 'Marte', 'Jupiter', 'Saturno', 'Urano', 'Neptuno']

nomCortos = [planeta for planeta in planetas if len (planeta) <= 6]
nomCortos
````

````python
#Otro ejemplo
cad = ['chb', 'ydb', 'thd', 'hgh']
nuevoCad = [c for c in cad 
if c.endswith('b') and c.startswith('c')]
nuevoCad
````

List comprehensions combinado con las funciones de min, max y sum dan soluciones de una línea.

````python
#Forma común
def numNegativos (nums):    #Se define la función numNegativos que tiene como parámetro a nums
  """ Regresa cuántos números negativos hay en la lista dada
  >>> numNegativos ([5,-5,10,-10, 15])
  2
  """
  nNegativos = 0
  for n in nums:
    if n < 0:
      nNegativos += 1
  return nNegativos
````

````python
numNegativos ([5,-5,10,-10, 15])
````

````python
def numN (nums):
  return len ([n for n in nums if n < 0])
````

````python
numN ([5,-5,10,-10, 15])
````

````python
#Otra forma 
def numNv1 (nums):
  return (sum([n < 0 for n in nums]))
````

````python
numNv1 ([5,-5,10,-10, 15])
````

##Iteración dentro de un diccionario
Se puede recorrer, de manera iterativa, a través de: <p>
* pares llave-valor: nombreDiccionario.items ()
* llaves: nombreDiccionario.keys()
* valores: nombrediccionario.values ()

Recorrido llaves-valor <p>
Se puede utilizar un for y nombreDiccionario.items ()

````python
#Ejemplo recorrido llaves-valor
usuario = {
    'userName' : 'uam azc',
    'host': 'kikikiki qui qui',
    'pwd': 'azc la mejor',
    
}
print ("Recorrido llave-valor")
for llave, valor in usuario.items ():
  print (f"Llave: {llave} Valor: {valor}", )
print ("~~~~~~~~~~~~~~~~~~~~~~~~~~")

````

````python
#Recorrido por todas las llaves
for k1 in usuario:
  print (k1)  
print ("~~~~~~~~~~~~~~~~~~~~~~~~~~")
#Equivalente a 
for k in usuario.keys ():
  print (k.upper())
print ("~~~~~~~~~~~~~~~~~~~~~~~~~~")

#Recorrido en orden por llaves
for k in sorted (usuario):
  print (k)
print ("~~~~~~~~~~~~~~~~~~~~~~~~~~")

for llave in usuario:
  print (llave, "\t---- ", usuario[llave])  
````

````python
#Recorrido por valores
print ("~~~~~~~~~~~VALORES~~~~~~~~~~~~~~~")  
for k in usuario.values():
  print (k)  
print ("~~~~~~~~~~~VALORES TITLE~~~~~~~~~~~~~~~")  
for v in usuario.values():
  print (v.title())   
````

````python
#Acceso a ciertas llaves
usuario = {
    'userName' : 'uam azc',
    'host': 'kikikiki qui qui',
    'pwd': 'azc la mejor',
    
}
datos = ['userName', 'host']
for k in usuario:
  if k in datos:
    valor = usuario[k].upper ()
    print (k, "==>", valor)
````

Considere el siguiente código que crea un diccionario
