# **Mientras**

Un ciclo `while` se ejecuta mientras una condición sea verdadera, por lo que los enunciados que abarca se pueden ejecutar 0 o más veces

```python
#Cuenta del 1 al 5
num = 1
while num <= 5:
  print (num)
  num += 1
```

## Uso de una bandera
Existen ocasiones donde varias condicones causan que un ciclo se detenga o continúe.

```python
prompt = "\nEscribe algo, y lo repetire para ti: "
prompt += "\nPara terminar escribe salir "
continuar = True
while continuar:
  mensaje = input (prompt)
  if mensaje == 'salir':
    continuar = False
  else:  print (mensaje)
print ("Gracias por utilizar este programa")
```

## break para salir de un ciclo

* La sentencia `break` hace que el ciclo se termine inmediatamente sin ejecutar el resto del código
* Puede usarse en cualquier ciclo

```python
#Break para salir de un ciclo
prompt = "\nEscribe las ciudades que has visitado: "
prompt += "\nPara terminar escribe salir "

while True:
  ciudad = input (prompt)
  if ciudad.lower () == 'salir': #para comparar minúsculas
    break
  else: print (ciudad)
print ("Continua viajando .....")
```

#**Para**

El `for` permite repetir un conjunto de sentencias un número determinado de veces (entero, por supuesto)

```python
#range produce una secuencia
x = 1
for i in range (5):
  print (x)
print ("Aquí estoy")  
```

La función range () tiene por default un valor de inicio en 0 y un incremento de 1, como tercer parámetro. 

```python
#Incremento de 3
for i in range (2, 30, 3):
  print (i)

  range
```

```python
#Recorrido de una lista con for
animales= ['lombriz','perro loco','perico','gallo','pollo']

for animal in animales:
  print (animal.title ())  #Inicia con mayúscula
```
