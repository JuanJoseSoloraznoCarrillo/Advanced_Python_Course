## Método constructor

Un método constructor es el encargado de inicializar los datos miembro de un objeto en particular. En Python la función que permite crear un objeto es __init__. <br>
Los métodos que inicial con __ son funciones especiales de Python que están conectadas con aspectos del lenguaje.<br>

* El método __init__ () se ejecuta automáticamente cuando se crea una nueva instancia
* Asegurarse que sean dos guiones antes y dos después, de lo contrario no se invocará
* El parámetro self se requiere en la definición del método y debe ser el primero
* Todos los métodos asociados con una instancia automáticamente pasan a self, que hace referencia a si mismo, por lo que en la declaración de un método siempre está presente y es el primero

### Primer ejemplo

Considerar un Auto, con datos miembro:

*   marca
*   modelo
*   año

Y funciones miembro:

*   __init__
*   descripción ()



````python
class Auto:
  def __init__ (self, ma, mo, a):
    self.marca = ma
    self.modelo = mo
    self.año = a

  def descripción (self):
    d = f'{self.año} {self.marca} {self.modelo}'
    return d

auto = Auto ('Porshe', 'Taycan Turbo s', 2021)
print (auto.descripción ()) #Acceso a un método miembro
print ("Marca: ", auto.marca) #Acceso al dato miembro
````

````python
# Se pueden agregar otros datos y funciones miembro
class Auto:
  def __init__ (self, ma, mo, a):
    self.marca = ma
    self.modelo = mo
    self.año = a
    self.odómetro = 0

  def descripción (self):
    d = f'{self.año} {self.marca} {self.modelo} {self.leeOdómetro ()}'
    return d

  def leeOdómetro(self):
    return str (self.odómetro) + ' km'  

  def actualizaOdómetro (self, d):
    if d > 0:
      self.odómetro += d

auto = Auto ('Porshe', 'Taycan Turbo s', 2021)
print (auto.descripción ()) #Acceso a un método miembro
print ("Marca: ", auto.marca) #Acceso al dato miembro
auto.actualizaOdómetro (3)
print (auto.descripción())
auto.actualizaOdómetro (30)
print (auto.descripción())
auto.actualizaOdómetro (-100)
print (auto.descripción())
````

````python
#Los atributos pueden modificarse directamente
auto.odómetro = 15
print (auto.descripción ())
````

##Métodos miembro privados

````python
class Amigo:
  def __init__ (self, nombre, apodo):
    self.nombre = nombre    # público
    self.__apodo = apodo    # privado

  def who (self):
    print ('nombre: ', self.nombre)
    print ('apodo:', self.__apodo)

  def __foo (self):   #método privado
    print ("Soy un método privado")

  def foo (self):     #método público
    print ('Soy un método público')
    self.__foo ()
````

````python
x = Amigo ('Roberto', 'Beto')
x.foo ()
````

````python
#No se puede
x.__foo ()
````

````python
#Se permite accesar un método privado
x._Amigo__foo ()
````

# Programación Orientada a Objetos - Herencia

Si la clase que se está diseñando es una versión especializada de otra existente, se puede usar la herencia. <p>

Es una relación **es un** entre clases. Permite derivar una clase más específica de otra más genérica. <p>

*Ejemplos:*
>* camioneta es una clase de vehículo usada para transportar objetos grandes
>* mamífero es una clase de animal con determinadas características que no todos los animales poseen.

La herencia es la capacidad de una clase de usar propidades y métodos de otra clase mientras añade su propia funcionalidad.

## Constructores, super y redefinición de métodos

Normalmente se inicializan todos los atribtos, incluyendo los de la clase padre <p>
En lugar de redefinir completamente un método también se puede invocar al método de la clase padre, a través de super (). <p>

Si el método de la clase padre no satisface los requerimientos de la clase hijo se puede definir un método en la clase hijo con el mismo nombre que el del padre, esto se conoce como override. Si no se desea redefinir todo, entonces puede invocarse al método de la clase padre a través de super (). <p>


````python
class Profesor (Persona):
  def __init__ (self, nombre, edad):
    super().__init__(nombre, edad)
    self.cursos = [] #inicializa un nuevo dato miembro

  def establecerOcupación (self):
    print ("Soy un profesor")

  def presentarse (self):
    super ().presentarse () #invoca a presentarse de persona
    self.establecerOcupación ()
    print ("Imparto", str (self.numCursos ()), "cursos")
    for curso in self.cursos:
      print (curso)
  
  def agregaCurso (self, curso):
    self.cursos.append (curso)

  def numCursos (self):
    return len (self.cursos)
````

````python
profe1 = Profesor ("Albus Dumbledore", 122)
profe1.agregaCurso ("Programación")
profe1.agregaCurso ("Data Mining")
profe1.presentarse ()
````

##Sobrecarga de operadores <p>

También se pueden sobrecargar operadores tales como +,-,*,/,<,>,in, len, entre otras, deben iniciar con doble guión bajo. Por ejemplo para '<' el método es: `__lt__` y para '>' `__gt__`.



````python
## Ejemplo
class Tweet:
    def __init__(self, mensaje, hora):
        self.mensaje = mensaje
        self.hora = hora 
        
    def __lt__(self, otro):
        return self.hora < otro.hora
        
    def __gt__(self, otro):
        return self.hora > otro.hora   
    

anteriorTweet = Tweet("Este es un tweet anterior",20)
recienteTweet = Tweet("Este es un tweet reciente",1000)
print(recienteTweet > anteriorTweet)
````

Estas funciones permiten que se puedan invocar automáticamente otras funciones que están preconstruidas como sorted ()

````python
tweets = [recienteTweet, anteriorTweet]

for tweet in sorted (tweets):
  print (tweet.mensaje)
````

La palabra clave `in` se usa para verificar elementos en una lista y llaves en un dicccionario. Por ejemplo:

````python
frutas = ['uva', 'manzana', 'cereza', 'durazno']
print ('cereza' in frutas)
print ('naranja' in frutas)
````
