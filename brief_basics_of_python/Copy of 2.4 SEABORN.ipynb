#**Swarm plot**

Swarm plot, beeswarm o gráfico de enjambre.

Se puede pensar que se tiene un conjunto de números, se ordenan y se pone un círculo por cada uno de los valores.

Por ejemplo: [1,7,4,2,1,4,1,2,5,4,2,7,3,1,4,5,1]
Ordenados [1,1,1,1,1, 2,2,2,  3, 4,4,4,4, 5,5, 7]

SwarmPlot.png


````python
#Para los pokemones se podría utilizar la paleta de colores
sns.swarmplot (x='Type 1', y = 'Attack', data = df, palette = pkmn_type_colors);
````
