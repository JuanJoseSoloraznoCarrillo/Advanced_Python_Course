## Funciones vectorizadas

Numpy ofrece versiones de las funciones estándar log, exp, sin, etc que trabajan sobre los elementos. <p>
También se conocen como ufuncs (*universal functions*)

```python
import numpy as np

#Ejemplo
z = np.array ([1,2,3])
np.sin (z)
```

Lo que elimina la necesidad de ciclos explícitos que trabajan elemento por elemento

```python
n = len (z)
y = np.empty(n)
for e in range(n):
  y[e] = np.sin (z[e])
y  
```

Las operaciones aritméticas básicas también trabajan así

```python
z
```

```python
(1/np.sqrt(2*np.pi))*np.exp(0.5*z**2)
```

No todas las funciones actúan elemento por elemento. <p>
Por ejemplo, la siguiente función causará un error

```python
def f(x):
  return 1 if x > 0 else 0
```

```python
x = np.random.randn(4)
x
```

```python
f(x)
```

La función `np.where` ofrece una versión alternativa

```python
np.where (x>0,1,0) #inserta 1 si x > 0, otro 0
```

También se puede usar la función `np.vectorize`.

```python
f = np.vectorize(f)
f(x)
```

##Comparaciones

Por lo general las comparaciones sobre los arreglos se realizan elemento por elemento.


```python
a = np.array ([2,4,6])
b = np.array ([2,4,6])
a == b
```

```python
a[0] = 10
a == b
```

```python
a != b
```

La situación es similar para >, <, >= y <=.<p>
También se pueden hacer comparaciones contra un escalar.

```python
r = np.linspace (0,10,5) #5 números entre 0 y 10 igualmente espaciados
r
```

```python
r > 3
```

```python
#útil para la extracción condicional
b = r > 3
b
```

```python
#como un solo paso
r[r>3]
```

```python
#2. Divida las ganacias de la tienda por trimestre
for i in año1:
  bim = np.reshape(año1,(4,3))

#De tal forma que las ganancias de enero se recuperan como año1[0][0] 
bim
```

```python
#3. Calcular las ganancias totales por trimestre
total = []
for t in bim:
  total.append(t.sum())

total
```
