#**JUPITER NOTEBOOKS**

#¿Qué es Jupyter Notebooks?

También conocido como "notebook", es un documento producido por la aplicación [Jupyter Notebook App](https://jupyter-notebook-beginner-guide.readthedocs.io/en/latest/what_is_jupyter.html#notebook-app) se caracteriza por tener celdas tanto de código (Python o R) como elementos de texto enriquecido. Además se puede tener ejecuciones, gráficas y otros recursos (*rich media*).

**snippets**

Este tipo de documentos son ideales para el trabajo de un científico de datos

#Jupyter Notebook App

Es una aplicación cliente-servidor que permite editar y ejecutar notebooks a través de un navegador web. También se puede instalar de manera local (por ejemplo [Anaconda](https://www.anaconda.com/products/individual))  y no requiere acceso a internet,   o instalarse en un servidor web remoto y luego acceder a ella a través de internet.

Sus dos componentes principales son:

* kernels
* dashboard




Un **kernel** es un "motor" computacional, que ejecuta el código contenido dentro del notebook. Cuando se abre un notebook, el kernel asociado se ejecuta de forma automática. Luego cuando se ejecuta un notebook (ya sea celda-por-celda o desde el menú Entorno de ejecución, el kernel ejecuta los cálculos computacionales y produce los resultados.

Por defecto la aplicación trae un kernel para el lenguaje Python, pero pueden agregarse kernels para otros lenguajes de programación como R.

El **dashboard** de la aplicación, es el componente gráfico que se visualiza cuando se ejecuta Jupyter Notebook App. Permite gestionar notebooks (agregar, crear, editar, etc), configurar y gestionar los Kernels disponibles (seleccionar, apagar, reiniciar, etc), entre otras opciones.

##¿Qué es una celda?

Un notebook se compone de un conjunto de celdas, unidad independiente que puede ser editada y ejecutada.

Existen dos tipos de celdas:

*   **Código**: contienen código que puede ser ejecutado, utilizando el kernel seleccionado (ej. Python, R)
*   **Texto**: contienen texto enriquecido (texto, tablas, figuras, etc.)

Cuando se abre un notebook, se carga la información de las celdas pero NO se ejecutan las celdas de código, para hacerlo se puede utilizar el menú de `Entorno de ejecución` o bien oprimir el botón de la izquierda

EntornoDeEjecución.png

Para **crear** una celda, simplemente oprimir el botón correspondiente al tipo que se desea.

Para **editar** una celda, basta con seleccionarla si es de código, o hacer doble click si es de Texto.

````python
def multiX2 (n):

  """
  Función definida por el programador
  multiplica por dos un número dado
  ocupa varias líneas
  """
  return n*2

print (multiX2 (5))  
````

````python


def describeMascota (tipoAnimal, nombreMascota):
  """Python debe mapear cada argumento en la llamada a la función"""
  """Comentario de una sola línea"""
  print ("Tengo una mascota, se llama ")
  print (nombreMascota, " y es un ", tipoAnimal, "\n")
````

````python
#En posicionales, el orden es importante
#De lo contrario se pueden obtener resultados inesperados, 
#erróneos o simpáticos
describeMascota("Baloo", "oso")

#Argumentos con clave
#Son pares nombre-valor que se pasan a la función, 
#donde la asociación se hace explícita
describeMascota (tipoAnimal="perro", nombreMascota="Bolt")
describeMascota (nombreMascota="Bagheera", tipoAnimal="pantera")
````

````python
help (describeMascota)
````

##Comandos mágicos

Como toda herramienta, existe una serie de comandos o funciones esenciales para trabajar. Para ver la lista de "comandos mágicos" disponibles en un kernel, se crea una celda de tipo "código" y se ejecuta el siguiente comando:

````python
%lsmagic
````

Para información sobre una función mágica en particular se usa `?` seguido de la función

````python
?%cd
````

````python
?%automagic
````

````python
%timeit c = [i ** 2 for i in range (1000)]
````

````python
%%timeit
  c2 = []
  for i in range (1000):
    c2.append (i**2)
````
