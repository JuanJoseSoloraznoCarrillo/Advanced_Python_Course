## Paso de argumentos

* La definición de una función puede tener varios parámetros
* La llamda a una función puede necesitar múltiples argumentos
* Se pueden pasar argumentos a las funciones de varias formas:


> * **posicional**, que necesita tner el mismo orden en el que se escribieron los parámetros
> * **con clave**, donde cada argumento consiste del nombre de la variable y su valor



````python
#Argumentos posicionales
#Python debe mapear cada argumento en la llamada a la función con los parámetros de la función. 
#La forma más simple es hacerlo en orden

def describeMascota (tipoAnimal, nombreMascota):
  print ("Tengo una mascota, se llama ")
  print (nombreMascota, " y es un ", tipoAnimal, "\n")
````

**Paso de parámetros por valor y por referencia**

Dependiendo del tipo de datos que se envían a una función se doferencían dos comportamientos:
* Paso por valor: se crea una copia local de la variable desntro de la función
* Paso por referencia: se maneja directamente la variable, los cambios realizados dentro de la función le afectarán también afuera

¿Cómo se pasan? <br>
* Tipos simples por valor: enteros, flotantes, cadenas, booleanos
* Tipos compuestos se referencia: listas, diccionarios, conjuntos
 

````python
#Simple por valor

#Paso por valor
def doble (num):
  num *=2
  print ("en doble n: ", num)
````

````python
i = 2
doble (i)
print ("En la llamada: ", i)
````

````python
#Simple por referencia

#Paso por referencia

#Paso por valor
def doble (num):
  num *=2
  print ("en doble n: ", num)
  return num

num = 10
num = doble (num)
print ("Aquí num vale: ", num)
````

Otro ejemplo, una función invoca a otra función

````python
def f1 (a):
  print ("En f1: ", a+1)

def f2 (b):
  f1 (b+5)
  print ("Regrese a f2: ", b)

f2 (40)  
c = 5
f2 (c)
print ("El valor de c es: ", c)
````

#**FUNCIONES DE ORDEN SUPERIOR**

## Funciones aplicadas a funciones

Las funciones que operan en otras funciones se conocen como de "orden superior" o de "alto nivel". <br>
Un ejemplo es la función **max**, que regresa el mayor de los argumentos, pero si se usa una función usando el argumento opcional key, regresa el argumento x que maximiza key(x). 

````python
def mod_5(x):
    """Regresa el residuo de dividir entre 5"""
    return x % 5

print(
    '¿Qué número es mayor?',
    max(100, 51, 14),
    '¿Qué número tiene el residuo mayor cuando se divide entre 5',
    max(100, 51, 14, key=mod_5),
    sep='\n',
)
````

Otro ejemplo


````python
def multPor5(x):
    return 5 * x

def invoca(fn, arg):
    """invoca fn con arg"""
    return fn(arg)

def dobleInvoca(fn, arg):
    """Lama fn con el resultado de llamar fn con arg"""
    return fn(fn(arg))

print(
    invoca(multPor5, 1),
    dobleInvoca(multPor5, 1), 
    sep='\n', 
)
````

Así, las funciones pueden tomar funciones como parámetros y devolver funciones como resultado.

La **programación funcional** promueve el uso de funciones para realizar cualquier tarea dentro de un programa, así un programa toma un conjunto de argumetos y regresa un resultado (*pure functions*).

Las tres operaciones fundamentales en la programación funcional son:

1. **Mapping**, aplica una trasformación
2. **Filtering**, aplica una función que satisface cierta condición
3. **Reducing**, aplcia una reducción y regresa un valor acumulado

La función **filter ()** permite procesar un iterable y extraer aquellos elementos que satisfacen una condición. Se aplica a un iterable y se regresa un objeto.

Son una alternativa a la list comprehension.

El primer argumento de filter () es una función que se pasa sin paréntesis, el segundo argumento un iterable, como lista, tupla o conjunto.

````python
def multiploDe5 (n):
   return n%5 == 0

números = [2, 5, 10, 21, 30, 47, 50]    

list (filter (multiploDe5, números))
````

````python
#Ejemplo con funciones lambda y definidas por el usuario

nums = [-5, -2, 0, 2, 5, 7]

#Usando una función lambda
positivos= filter (lambda n: n > 0, nums)
print (list (positivos))

#Usando una función definida
def esPositivo (n):
  return n > 0

list (filter (esPositivo, nums))  
````

Escribir una función para extraer los números primos en un intervalo dado, iniciar con una función que tome un entero como argumento y regrese `True` si el número es primo, `Falso` en otro caso.

````python
import math

def primo (n):
  if n <= 1:
    return False
  for x in range(2, int(math.sqrt (n))+1):
    if (n%x)==0:
      return False
  return True


print ("Es primo 5  --> ", primo (5))
print ("Es primo 12 --> ", primo (12))
print ("Es primo 17 --> ", primo (17))
````

````python
#Usar filter () para extraer los primos de un intevalo

positivos1 = (filter (lambda n: n%2==0, range (1, 80))) 

print (list (positivos1))
````
