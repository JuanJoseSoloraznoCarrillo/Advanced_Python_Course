# **REGRESIÓN LINEAL**

Es un método estadístico que permite construir una función de variables independientes (predictores) para predecir una variable dependiente (llamada respuesta).

Por ejemplo: dar o no un crédito bancario

Se llama regresión lineal **simple** cuando sólo hay una variable independiete y regresión lineal **múltiple** cuando hay más de una.

Dependiendo del contexto, a la variable modelada se le conoce como variable dependiente o varible respuesta, y a las variables independientes como regresores, predictores o *features*.


#Introducción

La **regresión** intenta explicar la relación entre distintas variables. Una de estas variables, llamada variable dependiente, es lo que se desea explicar usando una o más variables.

En la ***regresión lineal*** se asume que la variable dependiente puede expresarse aproximadamente como una combinación lineal de las independientes.

<img src = "https://i.ytimg.com/vi/zPG4NjIkCjc/maxresdefault.jpg" />



Otra forma de visualizarlo es el ajuste de una curva de datos observados.

##Pasos para establecer una regresión

Un ejemplo simple es predecir el peso de una persona cuando se conoce su estatura. Para hacerlo se necesita tener la relación entre la altura y el peso de una persona.

Pasos:

1. Llevar a cabo un experimento para obtener una muestra de valores observados de altura y su correspondiente peso
2. Crear una relación entre las variables
3. Encontrar los coeficientes del modelo creado y crear la ecuación matemática
4. Obtener las relaciones del modelo para conocer el error promedio en la predicción. También llamado residuales
5. Predecir los pesos de nuevas personas


La regresión lineal permite predecir una respuesta con una función lineal de predictores

$y = c_0 + c_1x_1 + c_2x_2 + ... + c_kx_k$
Donde $x_i$ son predictores y $y$ es la respuesta a predecir.

En una regresión lineal la relación entre las variables puede describirse a través de una línea recta. Las regresiones no-lineales producen líneas curvas,

Se puede aplicar regresión lineal con el paquete ***scikit-learn***

###Ecuación de la regresión lineal

Del álgebra elemental tenemos la ecuación de una recta:

$y = a + bx$

* $y$ es la variable dependiente

* $x$ es la variable independiente

* $b$ es la pendiente de la recta

* $a$ es el punto donde se atraviesa el eje y (intercept). Corresponde con el valor promedio de la variable respuesta $y$ cuando todos los predictores son cero

###Pasos para determinar si hay una relación entre variables



* Muchas veces queda a juicio del investigador  
* Siempre conviene hacer una gráfica para ver si los datos se ajustan a una línea antes de intentar encontrar la ecuación
* Pasos:

> 1. Realizar los cálculos de $\sum x$, $\sum y$, $\sum xy$, $\sum x^2$, $\sum y^2$

>2. Encontrar $a$ y $b$

$a = \frac{(\sum y)(\sum x^2)-(\sum x)(\sum xy)}{n((\sum x^2)-(\sum x)^2}$

$b = \frac{n(\sum xy) -(\sum x)(\sum y)}{n((\sum x^2)-(\sum x)^2}$
>3. Poner los valores en la ecuación

##**Encontrando Coeficientes óptimos**

Recordando el modelo de regresión lineal 

$y = c_0 + c_1x_1 + c_2x_2 + ... + c_kx_k + e$

Donde $x_i$ son predictores, $y$ es la respuesta a predecir y $e$ es el residuo o error.

$e$ es la diferencia entre el valor observado y el estimado por el modelo. Recoge el efecto de todas aquellas variables que influyen en $y$ pero no se incluyen en el modelo como predictores.

En la mayoría de los casos, los valores $c_i$ se desconocen y se obtienen estimaciones. Ajustar el modelo consiste en estimar, a partir de los datos disponibles, los valores de los coeficientes de regresión que maximizan la verosimilitud (*likelihood*), es decir, los que dan lugar al modelo que con mayor probabilidad puede haber generado los datos observados.





Una vez ajustado el modelo, es necesario verificar su utidad (calidad), en otras palabras, determinar qué tan bien se ajusta el modelo a los datos. Puede existir un gran error aunque sea la línea que mejor se ajusta a las observaciones dadas.

Medir la calidad del modelo permite mejorarlo iterativamente.

En muchas aplicaciones se mide la veracidad (*accuracy*) de la predicción. 

Primero se necesita resumir la calidad del modelo para que sea entendible. Por ejemplo si se comparan los valores predecidos con los valores actuales de 10,000 casas, se tendría una mezcla de buenas y malas predicciones. Lo más conveniente sería tener una sóla métrica, que resumiera las predicciones.

Algunas métricas son:


*   Error absoluto medio
*   Error cuadrático medio
*   Coeficiente de determinacion $R^2$.



errores.png

En general, un modelo se ajusta bien a los datos si las diferencias entre los valores observados y los valores de predicción del modelo son pequeñas y no presentan sesgo.

##Error absoluto medio *Mean Absolute error* (MAE)

El **error** para cada predicción es:

$error = actual-predicción$

Por ejemplo, si una casa cuesta `$150,000`  y la predicción es de `$100,000` el error es de `$50,000`.

Esta métrica toma el valor absoluto de cada error y luego obtiene el promedio.

#Error cuadrático medio *Mean squared error* (MSE)

En estadística, el error cuadrático medio (ECM), en inglés *Mean squared error (MSE)*, de un estimador mide el promedio de los errores al cuadrado, es decir, la diferencia entre el estimador y lo que se estima. 

Tiene las mismas unidades de medida que la cantidad que estima.

$MSE = \frac{1}{n}\sum_{i=1}^{n} {(yPred_i-y_i)}^2$

Este valor es dependiente de la muestra, ya que está calculado a partir de ella.

#Coeficiente de determinación

El coeficiente de determinación, también conocido como $R^2$ es una medida estadística de qué tan cerca están los datos de la línea de regresión.

Se define como el porcentaje de la variación en la variable de respuesta que es explicado por un modelo lineal. Es decir

$R^2 = \frac {variaciónExplicada} {variaciónTotal}$

Siempre está entre 0 y 1 (0 y 100%)

* 0 indica que el modelo no explica ninguna proporción de la variabilidad de los datos de respuesta en torno a su media
* 1 indica que le modelo explica toda la variablidad de los datos de respuesta en torno a su media.


CoeficienteDeDeterminación.png

El modelo de regresión de la izquierda explica aprox. el 40% de la varianza, mientras que el de la derecha cerca del 88%. Cuanto mayor sea la varianza explicada por el modelo de regresión, más cerca estarán los puntos de los datos de la línea de regresión ajustada. En teoría si un modelo explica el 100% de la varianza, los valores ajustados siempre serían iguales a los observados y por lo tanto sobre la línea de regresión.


#Ejemplo 1

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score

# Se lee el dataset diabetes
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)

#Se utiliza sólo la 1ra. característica
diabetes_X = diabetes_X[:, np.newaxis, 2]

# Los datos se dividen en entrenamiento y prueba
diabetes_X_train = diabetes_X[:-20]
diabetes_X_test = diabetes_X[-20:]

# El objetivo se divide en entrenamiento y prueba
diabetes_y_train = diabetes_y[:-20]
diabetes_y_test = diabetes_y[-20:]

# Creación del modelo de regresión lineal
regr = linear_model.LinearRegression()

# Entrenamiento del modelo usando los datos correspondientes
regr.fit(diabetes_X_train, diabetes_y_train)

# Se realizan las predicciones usando los datos de prueba
diabetes_y_pred = regr.predict(diabetes_X_test)

# Coeficientes
print('Coeficiente: \t', regr.coef_)
print ("Intercept:  \t", regr.intercept_)
# Error promedio al cuadrado
print('MSE: %.2f'
      % mean_squared_error(diabetes_y_test, diabetes_y_pred))
# Coeficiente de determinación: 1 predicción perfecta
print('Coeficiente de determinación: %.2f'
      % r2_score(diabetes_y_test, diabetes_y_pred))

# Graficación
plt.scatter(diabetes_X_test, diabetes_y_test,  color='black')

plt.xticks(())
plt.yticks(())

plt.show()
```

##"lineal"

Relación lineal (o asociación lineal) es un término estadístico que se utiliza para describir una relación lineal entre dos varibles. Las relaciones lineales pueden expresarse en forma gráfica, las variables deben estar conectadas por una línea recta o en formato matemático ($y = mx+b$).

El término "lineal" en los modelos de regresión hace referencia a que los parámetros se incorporan en la ecuación de forma lineal, no a que necesariamente la relación entre cada predictor y la variable respuesta tenga que seguir un patrón lineal.

Ejemplo de un modelo lineal donde el predictor $x_1$ es lineal respecto a $y$:

$y = c_0 + c_1x_1 + c_2log(x_1) + e$

Contraejemplo:

$y = c_0 + c_1x_1^{c_2} + e$

A veces, algunas relaciones no lineales pueden transformarse de forma que se puedan expresar de forma lineal:

$y = c_0 + x_1^{c_1} + e$

$log (y) = log (c_o) + c_1 log (x_1) + log(e)$

##Interpretación del modelo

Los principales elementos que pueden interpretarse en un modelo de regresión lineal son los coeficientes de los predictores.



*   $c_o$ ordenada en el origen o intercept, que corresponde con el valor esperado de la respuesta $y$ cuando todos los predictoresson cero.
*  $c_i$ coeficientes de regresión de cada predictor que indican el cambio promedio esperado de la variable respuesta $y$ al incrementar en una unidad de la varible predictora $x_i$, manteniéndose constantes el resto de las varibles.

La magnitud de cada coeficiente parcial de regresión depende de las unidades en las que se mida la variable predictora a la que corresponde, por lo que su magnitud no está asociada con la importancia de cada predictor.

Para poder determinar qué impacto tienen en el modelo cada una de las variables, se emplean los coeficientes parciales estandarizados, que se obtienen al estandarizar (sustraer la media y dividir entre la desviación estándar) las variables predictoras previo ajuste del modelo. En este caso,  $c_0$  se corresponde con el valor esperado de la variable respuesta cuando todos los predictores se encuentran en su valor promedio, y  $c_i$  el cambio promedio esperado de la variable respuesta al incrementar en una desviación estándar la variable predictora  $x_i$ , manteniéndose constantes el resto de variables.

Si bien los coeficientes de regresión suelen ser el primer objetivo de la interpretación de un modelo lineal, existen muchos otros aspectos (significancia del modelo en su conjunto, significancia de los predictores, condición de normalidad...). Estos últimos suelen ser tratados con poca detalle cuando el único objetivo del modelo es realizar predicciones, sin embargo, son muy relevantes si se quiere realizar inferencia, es decir, explicar las relaciones entre los predictores y la variable respuesta.

```python
from sklearn import metrics
from sklearn.metrics import r2_score
print ("Performance del modelo (conjunto de entrenamiento):")
yEntrenaPredicción = lm.predict(xEntrena)
print('MAE', metrics.mean_absolute_error(yEntrena, yEntrenaPredicción))
print('MSE', metrics.mean_squared_error(yEntrena, yEntrenaPredicción))
print('RMSE', np.sqrt(metrics.mean_squared_error(yEntrena, yEntrenaPredicción)))
print('R error al cuadrado', r2_score(yEntrena, yEntrenaPredicción))
print ('***********************')
print ("\nPerformance del modelo (conjunto de prueba):")
print('MAE', metrics.mean_absolute_error(yPrueba, predicción))
print('MSE', metrics.mean_squared_error(yPrueba, predicción))
print('RMSE', np.sqrt(metrics.mean_squared_error(yPrueba, predicción)))
print('R error al cuadrado', r2_score(yPrueba, predicción))
```

#VALORES FALTANTES

Los valores faltantes ¡suceden!

Hay que estar preparados para los dataset reales.

Muchas bibliotecas de ML (incluyendo scikit-learn) producen error cuando se intenta constuir un modelo si se tienen datos faltantes.

##3 enfoques:

**1. Borrar columnas con valores faltantes**

Es el enfoque más simple, a menos que falten muchos valores, el modelo perdera acceso a la información. Por otro lado, si se tienen muchos renglos, eliminar una columna con valores faltantes no se perderá mucho. 

En algunas ocasiones puede afectar los resultados, en otras no.




Dropout.png

**2. Imputación**

Rellenar los valores faltantes con algún número. Por ejemplo, promedio.


Imputation.png

**3. Extensión de la imputación**

La imputación es el enfoque estándar, y normalmente trabaja bien. Sin embargo, podrían estar arriba o abajo de los actuales, haber sólo un renglón con valores faltantes, etc. por lo que es conveniente analizar qué valores faltaban originalmente.

ExtImputation.png

```python
print (data.columns)
print (len (data.columns))
```

```python
from sklearn.model_selection import train_test_split

# Separar objetivo y características
y = data.Price
X = data.drop(['Price'], axis=1)

# Dividir datos en subconjuntos de entrenamiento y validación 
X_train_full, X_valid_full, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2,
                                                                random_state=0)
```

```python
# Eliminar columnas con valores faltantes (enfoque simple)
cols_with_missing = [col for col in X_train_full.columns if X_train_full[col].isnull().any()] 
X_train_full.drop(cols_with_missing, axis=1, inplace=True)
X_valid_full.drop(cols_with_missing, axis=1, inplace=True)
```

```python
# "Cardinalidad" número de valores únicos en una columna
# Selecciona columnas categóricas con baja cardinalidad (arbirario)
bajaCardinalidadCols = [cnom for cnom in X_train_full.columns if 
                        X_train_full[cnom].nunique() < 10 and 
                        X_train_full[cnom].dtype == "object"]

# Selecciona columnas numéricas
numericasCols = [cnom for cnom in X_train_full.columns if 
                 X_train_full[cnom].dtype in ['int64', 'float64']]

# Guarda columnas seleccionadas
my_cols = bajaCardinalidadCols + numericasCols
X_train = X_train_full[my_cols].copy()
X_valid = X_valid_full[my_cols].copy()
```

```python
X_train.head ()
```

```python
#Lista de las variables categóricas (en este caso texto)
s = (X_train.dtypes == 'object')
object_cols = list(s[s].index)

print ("Variables categóricas: ")
print (object_cols)
```

Se definirá una función para medir la calidad de cada enfoque que trata con las variables categóricas. Reporta a través del mean absolute error (MAE), mientras más bajo mejor.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

# Función para comparar los tres enfoques
def score_dataset(X_train, X_valid, y_train, y_valid):
    model = LinearRegression()
    model.fit(X_train, y_train)
    preds = model.predict(X_valid)
    return mean_absolute_error(y_valid, preds)
```

Score con el Enfoque 1 (Eliminar variables categóricas)

```python
drop_X_train = X_train.select_dtypes(exclude=['object'])
drop_X_valid = X_valid.select_dtypes(exclude=['object'])

print("MAE con enfoque 1 (Eliminar variables categóricas):")
print(score_dataset(drop_X_train, drop_X_valid, y_train, y_valid))
```

Score con el Enfoque 2 (Codificación ordinal)

```python
from sklearn.preprocessing import OrdinalEncoder

# Copia de los originales
label_X_train = X_train.copy()
label_X_valid = X_valid.copy()

# Aplicar codifación ordinal a datos categóricos
ordinal_encoder = OrdinalEncoder()
label_X_train[object_cols] = ordinal_encoder.fit_transform(X_train[object_cols])
label_X_valid[object_cols] = ordinal_encoder.transform(X_valid[object_cols])

print("MAE con enfoque 2 (Codificación ordinal):") 
print(score_dataset(label_X_train, label_X_valid, y_train, y_valid))
```

Score con el Enfoque 3 (Codificación One-Hot)

```python
from sklearn.preprocessing import OneHotEncoder

# Aplicar one-hot encoder a cada columna con datos categóricos
OH_encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)
OH_cols_train = pd.DataFrame(OH_encoder.fit_transform(X_train[object_cols]))
OH_cols_valid = pd.DataFrame(OH_encoder.transform(X_valid[object_cols]))

# Eliminar el index 
OH_cols_train.index = X_train.index
OH_cols_valid.index = X_valid.index

# Eliminar columnas categóricas 
num_X_train = X_train.drop(object_cols, axis=1)
num_X_valid = X_valid.drop(object_cols, axis=1)

# Insertar columnas one-hot encoded columns 
OH_X_train = pd.concat([num_X_train, OH_cols_train], axis=1)
OH_X_valid = pd.concat([num_X_valid, OH_cols_valid], axis=1)

print("MAE con enfoque 3 (One-Hot Encoding):") 
print(score_dataset(OH_X_train, OH_X_valid, y_train, y_valid))
```

**¿Qué enfoque es mejor?**

Por lo general el enfoque 1 tiene un desempeño más pobre que el enfoque 3
