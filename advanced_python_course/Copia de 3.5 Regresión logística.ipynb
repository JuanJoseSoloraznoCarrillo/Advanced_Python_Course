# **REGRESIÓN LOGÍSTICA**

Se usa para predecir la probabilidad de ocurrencia de un evento ajustando los datos a una curva logística. 

Un modelo logístico de regresión se construye usando la siguiente ecuación

$logit (y) = c_0 + c_1x_1 + c_2x_2 + ... + c_kx_k$

donde:
* $x_1$, $x_2$, ... , $x_k$ son predictores, 
* $y$ es la respuesta de predecir $logit(y) = ln (\frac {y}{1-y}) = log(p)-log(1-p)$

Los predictores pueden ser categóricos, continuos o ambos

Un ejemplo clásico en ML es la clasificación de correos: dado un conjunto de atributos para cada correo, tales como número de palabas, ligas, dibujos, determinadas palabras, etc, predicir cuando es spam o no.

Otros ejemplos:

* clasificación de películas: entretenimiento, aventura, etc.
* detección de una enfermedad
* abandono de una compra online

La función logística también es llamada función sigmoide, que se describe reescribiendo la fórmula anterior como:

$f(x) = \frac {1}{1+e^{-x}}$

una curva que regresa `uno` si el dominio tiende a infinito, o regresa `cero` si tiende a infinito negativo.

#**Regresión lineal vs. regresión logística**

La regresión lineal es un algoritmo de aprendizaje supervisado que predice valores continuos, asume que existe una relación lineal entre las variables dependientes e independientes. Es decir encuentra la mejor línea/plano que describe dos o más variables.

Por otra parte la regresión logística es otro algoritmo de aprendizaje de máquina supervisado que aopya la clasificación binaria.

Aunque son diferentes, matemáticamente con pasos adicionales la regresión lineal puede convertirse en regresión logística.

##**Pasos en Regresión Lineal**

**Paso 1**

Consideremos un dataset donde $x$ es una variable independiente y $Y$ es una función de $x$ ($Y=f(x)$). Usando regresión lineal se formula la siguiente ecuación:

$Y = mx+c$

Ecuación de la línea recta, donde $m$ es la pendiente de la línea y $c$ es el intercept.

**Paso 2**

Para derivar la línea que mejor se ajusta, se asignan valores aleatorios a $m$ y $c$ para calcular el valor de $Y$ para $x$, $Y$ es el valor de salida.

**Paso 3**

Como es un algoritmo de ML supervisado, se conoce el valor actual $Y$ (variable dependiente). Representemos el valor de salida calculado como $\widehat y$. Se puede verificar que tan adecuada o no es la predicción.

En el caso de la regresión lineal, este error (residual) se peuede calcular usando el método MSE (*mean squared error*) que se nombra como la función ***loss***.

$L = \frac {1}{n} \sum {(y-\widehat y)}^2$

donde $n$ es el número de observaciones.


**Paso 4**

Para obtener la mejor línea ajustada, se debe minimizar la función loss. Una forma común de hacerlo es a través de la técnica del **gradiente descendente**.

*Gradiente descendente*

El gradiente es el conjunto de todas las derivadas parciales de una función. 

Si se grafica la función loss para $m$ y $c$, se verá una curva. Conforme se minimice la función de loss, se alcanzará la parte más baja de la curva

El gradiente es un vector que indica en qué dirección necesitan ajustarse los pesos, ie., cómo cambian los pesos para hacer que la función loss cambie rápidamente.

GradienteDesEst1.png

**Paso 5**

Una vez que la función loss se ha minimizado, se obtiene la ecuación final para el mejor ajuste de la línea y se puede predecir el valor de $Y$ para una $x$ dada.

##**Pasos en Regresión Logística**

Fundamentalmente la regresión logística se usa para clasificar elementos de un conjunto en dos grupos (clasificación binaria) a través de la probabilidad de cada elemento en el conjunto.

Se utiliza un umbral, si la probabilidad de un elemento es mayor que ese valor entonces se clasifica en un grupo o vice versa.

**Paso 1**

Para calcular la separación binaria, primero, se determina la línea mejor ajustada siguiendo los pasos de la Regresión lineal.

**Paso 2**

La línea de regresión obtenida es susceptible a valores atípicos, lo que podría no dar bueos resultados.

Así, los valores predecidos se convierten en probabilidades para la función sigmoide.

La ecuación de la sigmoide:

$S(x) = \frac {1}{1+e^{-x}}$

Dado un número real, la función sigmoide regresará un valor entre 0 y 1.

Así, si se alimenta un valor de salida $\widehat y$ a la función sigmoide regresará un valor de probabilidad entre 0 y 1.

**Paso 3**

Finalmente, el valor de la salida de la función sigmoide se convertirá a 0 ó 1 (valores discretos) dependiendo del umbral. El umbral normalmente se selecciona en 0.5

```python
X = df[['gmat', 'gpa', 'experiencia']]
 #predictores o características 
y = df['admitido'] #Separación en entrenamiento y prueba 
XEntrena, XPrueba, yEntrena, yPrueba = train_test_split(X,y, test_size=0.25, random_state=0) 
#Aplicación de la regresión logísitca 
regLog = LogisticRegression() 
regLog.fit(XEntrena, yEntrena) 
yPred=regLog.predict(XPrueba) #Matriz de confusión 
matrizConf=pd.crosstab(yPrueba, yPred, rownames = ['Actual'], colnames=['Predicción'])
sn.heatmap(matrizConf, annot=True)
print("Accurancy: ", metrics.accuracy_score(yPrueba, yPred))
```

#**Probabilidades condicionales**

#1. Eventos y probabilidades

Una contingencia es algo que puede o no suceder. Los eventos son conjuntos de contingencias a las que se asocian probabilidades de ocurrencia. Un evento ocurrirá cando suceda alguna de las contingencias que lo integran.

Como son conjuntos, los eventos pueden representarse a través de la teoría de conjuntos.

Por ejemplo:

* $A \bigcup B$ es el evento que consiste en que ocurra el evento $A$ o el evento $B$.

* $A \bigcap B$ es el evento que consiste en que ocurran ambos eventos $A$ y $B$.


$\Omega$ incluye todas las posibles ocurrencias

$\emptyset$	 no incluye ningúna ocurrencia

$\Omega$ \\ $A$ el complemento de $A$, significa que no ocurra $A$, también se usa $\widehat A$.

La **probabilidad** es una función que asigna a conjuntos (eventos) un número entre 0 y 1. Para cada evento $A$, $P(A) = x, 0 \leq x \leq 1$

La función $p$ está sujeta a los axiomas de Kolmogorov:

* $P(A) \geq 0$
* $P(\Omega) = 1$
* $P(\bigcup A_i) = \sum_i P(A_i$) si los $A_i$ son eventos *mutuamente excluyentes*, es decir su intersección es vacía.

De estos axiomas de probabilidad se deducen propiedades razonables, p.e., si $A \subset B$, entonces $P(A) \leq P(B)$

Algunos no encuentran la definición anterior satisfactoria, por lo que definen probabilidad como: 

*la máxima cantidad de dinero que apostarías si alguien te ofreciese un euro si el evneto en cuestión sucediese.*

> **Probabilidad condicional**: Probabilidad de uno (o más) eventos dada la ocurrencia de otro evento, p.e., $P(A$ dado $B)$ o $P(A|B)$

#Teorema de Bayes

El Teorema de Bayes proporciona la forma de calcular una probabilidad condicional y una forma de pensar sobre la selección y evaluación de diferentes modelos para un conjunto de datos dado en ML.

El modelo planteado al inicio de esta sección partía del clima y, en función de él, definía la probabilidad de tráfico. 

Es posible preguntarse cuál sería la probabilidad de que estuviese lloviendo a partir de la información del estado del tráfico. Es decr, calcular, $P(lluvia|tráfico)$ y en general, $P(T|A)$, ie., invertir las propiedades condicionales.

Para contestar la pregunta anterior:

$P(T|A) = \frac {P(T,A)} {P(A)}$

y tanto $P(T,A)$ como P(A) son conocidos. Desarrollando

$P(T|A) = \frac {P(T,A)} {P(A)} = \frac {P(A|T)P(T)} {P(A)}$

La expresión anterior se conoce como **Teorema de Bayes**.

Para nuestro modelo

$P(lluvia|tráfico) = \frac {P(tráfico|lluvia)P(lluvia)} {P(tráfico)} = \frac {0.4 x 0.1)}{0.13} \thickapprox 0.31$

En general, llueve el 10% de los días; con la información adicional de que hay tráfico, la probabilidad estimada de lluvia crece hasta el 31%.

Esto es útil, ya que normalmente solo conocemos las probabilidades condicionales menos interesantes.

## **Clasificador Naive Bayes**

El Teorema de Bayes es la base para el razonamiento deductivo, que se centra en determinar la probabilidad de que ocurra un evento basándose en el conocimiento previo de las condiciones que podrían estar relacionadas con el evento.

> La clasificación es una forma de aprendizaje supervisado.

> Todos los puntos que tienen la misma etiqueta pertenecen a la misma clase

> Puede haber más de una etiqueta

> Es un buen clasificador pero no predictor

La **clasificación Naive Bayes** es un método simple y rápido.

Su velocidad se debe a simplificaciones sobre la distribución de probabilidad de las características independientes.

Es especialmente útil cuando no hay suficientes características

Suponga que cada etiqueta L tiene una combinación de características con cierta probabilidad

𝑃(𝑐𝑎𝑟𝑎𝑐𝑡𝑒𝑟í𝑠𝑡𝑖𝑐𝑎𝑠│𝐿)


La idea principal de la clasificación Bayesiana es cambiar la dirección de la dependencia: se desea predecir la etiqueta basada en las características

$𝑃(𝐿│𝑐𝑎𝑟𝑎𝑐𝑡𝑒𝑟í𝑠𝑡𝑖𝑐𝑎𝑠)=\frac {𝑃(𝑐𝑎𝑟𝑎𝑐𝑡𝑒𝑟í𝑠𝑡𝑖𝑐𝑎𝑠│𝐿)𝑃(𝐿)}{𝑃( 𝑐𝑎𝑟𝑎𝑐𝑡𝑒𝑟í𝑠𝑡𝑖𝑐𝑎𝑠)}$

Por el teorema de Bayes


Asumamos dos etiquetas $L_1$ y $L_2$ y sus distribuciones asociadas: 𝑃(𝑐𝑎𝑟𝑎𝑐𝑡𝑒𝑟í𝑠𝑡𝑖𝑐𝑎𝑠│𝐿1)  𝑦 𝑃(𝑐𝑎𝑟𝑎𝑐𝑡𝑒𝑟í𝑠𝑡𝑖𝑐𝑎𝑠│𝐿2). Si se tiene un punto de datos con características, cuyos datos no se conocen, se podría usar:

$ \frac {𝑃(𝐿_1│𝑐𝑎𝑟𝑎𝑐𝑡𝑒𝑟í𝑠𝑡𝑖𝑐𝑎𝑠)}{𝑃(𝐿_2│𝑐𝑎𝑟𝑎𝑐𝑡𝑒𝑟í𝑠𝑡𝑖𝑐𝑎𝑠)}= \frac {𝑃(𝑐𝑎𝑟𝑎𝑐𝑡𝑒𝑟í𝑠𝑡𝑖𝑐𝑎𝑠│𝐿_1)𝑃(𝐿_1)}{𝑃(𝑐𝑎𝑟𝑎𝑐𝑡𝑒𝑟í𝑠𝑡𝑖𝑐𝑎𝑠│𝐿_2)𝑃(𝐿_2)} $

Si es mayor que 0, la etiqueta sería $L_1$, en caso contrario $L_2$
Las probabilidades a priori $P(L_1)$ y $P(L_2)$ pueden obtenerse de los datos de entrada.


La librería scikit-learn contiene esta familia de algoritmos
[Naive Bayes](https://scikit-learn.org/stable/modules/naive_bayes.html)

En el algoritmo Gaussian Naive Bayes, los valores continuos están asociados con cada característica y se asumen que tienen una distribución Gausiana (también llamada normal). Cuando se grafica la curva tiene una forma de campana



##Ejemplo **Naive Bayes Breast cancer**

```python
#Lectura de la Base de datos
from sklearn import datasets
dataset = datasets.load_breast_cancer ()
```

Características de la base de datos

```python
print ("Keys del dataset\n")
print (dataset.keys ())
print ("Descripción:\n")
print (dataset.DESCR)
```

```python
#Selección de todas las columnas
X = dataset.data
#Etiquetas
y = dataset.target
#Construcción de los conjuntos de entrenamiento y prueba
from sklearn.model_selection import train_test_split
XEntrena,XPrueba,yEntrena,yPrueba = train_test_split (X, y, test_size = 0.2)
```

Construcción del modelo Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
modelo = GaussianNB ()
modelo.fit (XEntrena, yEntrena)
yPred = modelo.predict (XPrueba)
```

```python
#Evaluación del modelo
from sklearn.metrics import confusion_matrix
matriz = confusion_matrix(yPrueba, yPred, labels=[0,1])
print ("Matriz de confusión")
print (matriz)
from sklearn.metrics import precision_score
prec = precision_score(yPrueba, yPred)
print ("Precisión del modelo: ", prec)
```

Escribir una función plantaClasificación que haga lo siguiente:

Lea la base de datos de iris de sklearn (sklearn.datasets.load_iris)

Separe los datos en entrenamiento y clasificación usando train_test_split en una proporción 80-20 con random_state=0 para hacerlo determinístico

Usar para entrenamiento un ajuste Naive Bayes Gaussiano

Predecir las etiquetas de los datos de prueba

```python
# lee el data set iris
from sklearn.datasets import load_iris
iris = load_iris()
  
# almacena las características den una matriz (X) y la respuesta en el vector (y)
X = iris.data
y = iris.target
  
# separa en conjuntos de entrenamiento y prueba
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
  
# Se entrena al modelo
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train)
  
# se realizan predicciones con el conjunto de prueba
y_pred = gnb.predict(X_test)
  
# comparación de los valores respuesta con las predicciones
from sklearn import metrics
print("Gaussian Naive Bayes model accuracy(en %):", metrics.accuracy_score(y_test, y_pred)*100)
```
