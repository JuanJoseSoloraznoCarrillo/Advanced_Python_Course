* Cree una lista de personas para invitarlas a comer (cuando termine la pandemia) <br>
Use esa lista para imprimir la invitación personalizada de cada persona, usando el mismo texto.

* El tercer invitado le dice que no puede asistir, por lo que deberá remplazarlo con un nuevo invitado<br>
Imprima la invitación para este nuevo invitado

* Le acaban de regalar una mesa más grande por lo que decide invitar a más personas, inserte a tres nuevos invitados, inserte a uno al principio, uno al final y a uno enmedio  <br>

* La cuarentena a regresado y solo puede tener dos invitados, use pop () para eliminar de la lista a todos los invitados hasta que solo queden dos. Envíe mensaje a todos los que no podrán asistir.  <br>

```python
invitados = ["persona_1","persona_2","persona_3"]

for i in range(len(invitados)):
  print("Invitacion para: ",invitados[i])

del invitados[2]
invitados.insert(2,"new_person")

invitados.append("last_person")
invitados.insert(0,"first_person")
print("\n")

for i in range(len(invitados)-2):
  print("No podra asistir",invitados.pop())
  
print("\n")
for i in range(len(invitados)):
  print("Sí asistirá: ",invitados[i])
```

```python
import random
Numeros = list(range(30,61,7))
random.shuffle (Numeros)
print(Numeros)

print(max(Numeros))
print(min(Numeros))
print(sum(Numeros))

a = int((len(Numeros)-1)/2)
print(Numeros[a])

Numeros.sort()

NumerosOrd = reversed(Numeros)

cuadrados = []
for num in NumerosOrd:
  numAlCuadrado = num **2
  cuadrados.append (numAlCuadrado)
print (cuadrados)
```
