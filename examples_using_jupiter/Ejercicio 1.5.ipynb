Escribir una función que acepte una lista de ingredientes que desea una persona para su sándwich. La función debe tener un parámetro que agrupa todos los ingredientes y debe imprimir el resumen de cómo se elaborará.

```python
def  sándwich(*ingredientes, temp = "caliente"):
  '''Eleboración de sándwich  '''
  print ("\nHacer sándwich", temp, "con los siguientes ingredientes:")
  for ingrediente in ingredientes:
    print (f"-{ingrediente}")
```

Llamar a la función tres veces usando un número diferente de argumentos a la vez.

```python
ingredientespara = ['huevo','queso', 'espinacas', 'jamón','jamón','jamon','jamón']

sándwich('panela', temp = 'frio')
sándwich('queso', 'espinacas', 'jamón',temp = 'frio')
sándwich('queso', 'espinacas')
sándwich(*ingredientespara, temp = 'frio')
```

##Ejercicio

Escribir una función para extraer los números primos en un intervalo dado, iniciar con una función que tome un entero como argumento y regrese `True` si el número es primo, `Falso` en otro caso.

```python
import math

def primo (n):
  if n <= 1:
    return False
  for x in range(2, int(math.sqrt (n))+1):
    if (n%x)==0:
      return False
  return True


print ("Es primo 5  --> ", primo (2))
print ("Es primo 12 --> ", primo (12))
print ("Es primo 17 --> ", primo (17))
```

```python
#Usar filter () para extraer los primos de un intevalo

positivos1 = (filter (lambda n: n%2==0, range (1, 80))) 

print (list (positivos1))
```

```python
list (filter (primo, range (1, 80)))
```

```python
def palindrome (pal):
  if pal == pal[::-1]:
    return True
  else:
    return False
    
print (palindrome ("racecar"))
print (palindrome ("Python"))
```

Determinar qué palabras son palindroma

```python
palabras = ("filtro", "ana", "hola", "madam", "racecar", "anita lava la tina")

list(filter(palindrome,palabras))
```

```python
def palindrome_2 (pal):
  pal= pal.replace(" ","")
  if pal == pal[::-1]:
    return True
  else:
    return False

palabras = ("filtro", "ana", "hola", "madam", "racecar", "anita lava la tina")
list(filter(palindrome_2, palabras))
```

```python
def palindrome_2 (pal):
  pal= pal.replace(" ","")
  reversa = "".join(reversed(pal))
  return pal == reversa
  
list(filter(palindrome_2, palabras))
```
