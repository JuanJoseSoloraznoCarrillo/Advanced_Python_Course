*Ejercicio:* ¿Qué sucede si el número de colores es menor que el número de datos?

¿Cómo podría dibujarse la gráfica anterior con sólo 3 colores de puntos?

```python
#De colores y tamaños
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colores = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])
tamaños = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

a = plt.subplots(1, 3, figsize=(15,5)) #tamaños cantidad de gráficas

a = plt.subplot(1,3,1) #(nrenglones, ncolulmnas, índice)
a.title.set_text('Original')
plt.scatter(x, y, c=colores, s = tamaños);


#Ejercicio con solo tres colores

colores_1 = np.random.randint(4, size=(13))
tamaños_1 = np.random.randint(500, size=(13))
b = plt.subplot (1,3,2)
b.title.set_text("Prueba")
plt.scatter(x, y, c=colores_1, s = tamaños_1);


#Otra forma de hacerlo:

color_2 = np.array(["black","red","blue"])
colores_2 = np.random.choice(color_2, size=13)
d = plt.subplot (1,3,3)
d.title.set_text("Prueba 2")
plt.scatter(x, y, c=colores_2, s = tamaños);
```
