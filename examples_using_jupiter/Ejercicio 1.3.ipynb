##Ejercicio

En lugar de imprimir la última ciudad, guardarla en una lista, no debe haber repetidas y cuando el usuario haya terminado deberá imprimarse el número de ciudades visitadas y el nombre de cada una. </br> </br>
ciudades = [] # crea una lista vacía <br>
ciudades.append (ciudad) #agrega ciudad a ciudades <br>
if ciudad.lower () == 'salir': continue   #para controlar el while

```python
prompt = "\nEscribe las ciudades que has visitado: "
prompt += "\nPara terminar escribe salir "

ciudades = [] 

while True:
  ciudad = input(prompt)
  if ciudad.lower() == 'salir': #para comparar minúsculas
    break
  else: 
    if ciudad in ciudades:
      print("Esa ciudad ya esta en la lista....")
    else: ciudades.append (ciudad)

cont = 0

print("\nTotal de ciudades visitadas =", len(ciudades))
for ciudad in ciudades:
  cont += 1
  print("Ciudades visitadas ----> {}-. {}".format(cont,ciudad))
```

Calcular la suma de los dígitos de un número dado por el usuario. <br>
Ejemplos:


*   Entrada: 123        Salida: 6
*   Entrada 12345       Salida: 15


```python
prompt = "\nIngrese el número entero: "
prompt += "\nPara terminar escribe salir "

Numeros = []

while True:
  Numero = input(prompt)
  if Numero.lower () == 'salir': #para comparar minúsculas
    break
  else: 
    N = int(Numero)
    Numeros.append(N)
  
def sum(x):
  '''Suma los numeros enteros de una lista'''
  suma = 0
  for i in x:
    suma += i
  return suma

sum(Numeros)
```

Un número de 3 dígitos es llamado número de **Armstrong** si la suma de los cubos de sus dígitos es igual al mismo número. <br>
Ejemplo: <br>
$ 153$ ya que $1^3 + 5 ^3 + 3^3 = 1 + 125 + 27 = 153$ <br>
Escribir un programa con todos los números de Armstrong entre 100 y 500

```python
def Armstrong(min,max):
  '''identifica los números de Armstrong dentro de los límites dados y los guarda en una lista.'''
  NumArmtrong = []
  for n in range(min,max+1):
    A = str(n)
    aux = 0
    for i in A:
      num = int(i)
      aux += num**3
    if aux == n:
      NumArmtrong.append(aux)
  return NumArmtrong

Armstrong(100,500)
```
