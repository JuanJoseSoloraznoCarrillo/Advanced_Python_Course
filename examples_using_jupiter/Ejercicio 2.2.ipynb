```python
import matplotlib.pyplot as plt
import numpy as np

#Fuente 
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
font1 = {'family':'serif','color':'red','size':10}
font2 = {'family':'serif','color':'darkred','size':12}
#Posición del título
plt.title("Datos de entrenamiento", fontdict = font1, loc = 'right')
plt.xlabel("Pulso promedio", fontdict = font2)
plt.ylabel("Calorias quemadas", fontdict = font2)

plt.plot(x, y);
```

```python
#Ejercicio: Cambiar el título a la izquierda (left) de la gráfica anterior
plt.title("Datos de entrenamiento", fontdict = font1, loc = 'left')
plt.xlabel("Pulso promedio", fontdict = font2)
plt.ylabel("Calorias quemadas", fontdict = font2)

plt.plot(x, y);
```

##Ejercicio

Hacer una gráfica lo más parecido a la siguiente

```python
x_ = np.linspace(0,10)
plt.plot(x_, np.sin(x_))
plt.plot(x_, np.cos(x_))
plt.title ("Sine and Cosine");
plt.xlabel("x axis label");
plt.ylabel("y axis label");
plt.legend(["Sine","Cosine"],loc = "lower left")
```

```python
#Ejercicio, cambiar las líneas del grid (axis = 'y')

#propiedades del grid
plt.plot(x, y,color='black',linewidth= 6)
plt.title ("Datos de entrenamiento")
plt.xlabel("Pulso promedio")
plt.ylabel("Calorias quemadas")
plt.grid(axis = 'x',color = 'blue', linestyle = '--', linewidth  = 2)
plt.grid(axis = 'y',color = 'green', linestyle = '-', linewidth  = 0.5)
```

```python
#De colores y tamaños
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colores = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])
tamaños = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

a = plt.subplots(1, 3, figsize=(15,5)) #tamaños cantidad de gráficas

a = plt.subplot(1,3,1) #(nrenglones, ncolulmnas, índice)
a.title.set_text('Original')
plt.scatter(x, y, c=colores, s = tamaños);


#Ejercicio con solo tres colores

colores_1 = np.random.randint(4, size=(13))
tamaños_1 = np.random.randint(500, size=(13))
b = plt.subplot (1,3,2)
b.title.set_text("Prueba")
plt.scatter(x, y, c=colores_1, s = tamaños_1);


#Otra forma de hacerlo:

color_2 = np.array(["black","red","blue"])
colores_2 = np.random.choice(color_2, size=13)
d = plt.subplot (1,3,3)
d.title.set_text("Prueba 2")
plt.scatter(x, y, c=colores_2, s = tamaños);
```
